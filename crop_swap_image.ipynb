```python
import cv2
import matplotlib.pyplot as plt
```

```python
#loading the image
iron_man=cv2.imread('ironman.jpg')
#resizing the image
iron_man=cv2.resize(iron_man,(1000,1000))
```

```python
#let's have a look on ironman
plt.imshow(iron_man)
```

```python
#cropping face
iron_man_crop=iron_man[80:400,300:800]
```

```python
plt.imshow(iron_man_crop)
```

```python
#loading the image
captain_america=cv2.imread('captainamerica.jpg')
#resizing the image
captain_america=cv2.resize(captain_america,(1000,1000))
```

```python
#let's have a look on captain america
plt.imshow(captain_america)
```

```python
#cropping face
captain_america_crop=captain_america[80:400,300:800]
```

```python
plt.imshow(captain_america_crop)
```

```python
#swapping the face of captain america with iron man.
iron_man[80:400,300:800]=captain_america_crop
```

```python
plt.imshow(iron_man)
```
