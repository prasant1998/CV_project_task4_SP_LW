```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#loading the image
iron_man=cv2.imread('iron_man.jpg')
#resizing the image
iron_man=cv2.resize(iron_man,(1000,1000))
```

```python
#let's have a look on ironman
plt.imshow(iron_man)
```

```python
#loading the image
thor=cv2.imread('thor.jpg')
#resizing the image
thor=cv2.resize(thor,(1000,1000))
```

```python
#let's have a look on thor
plt.imshow(thor)
```

```python
#arranging both images vertically
avengers_vertical=np.vstack([iron_man,thor])
```

```python
plt.imshow(avengers_vertical)
```

```python
#arranging both images horizontally
avengers_horizontal=np.hstack([iron_man,thor])
```

```python
plt.imshow(avengers_horizontal)
```
